### Approaching all links

Now that test is successful, we approach to get all avaliable links, starting from getting the `qnaics` code. `qnaics` code are 4 digits code for quick search. We will try other code later. That content is probably same.

```python
url_index='https://accessnc.nccommerce.com/business/business_custom_search_infogroup.html'
data_all_codes=requests.get(url_index,headers=set_headers)
```

Store the codes for quick access

```python
data_all_codes_parse=bs(data_all_codes.text,'html.parser')
with open("qnaics_codes.txt","w") as f:
    for line in data_all_codes_parse.find_all('option'):
        if(line.text.strip().__len__()>4 or line.text.strip().__len__()<3):
            continue
        f.write(line.text+"\n")
```

Now, start getting all links

```python
import requests
import os
import time
from bs4 import BeautifulSoup as bs
from tqdm.notebook import tqdm

# Read NAICS codes from the file
with open("qnaics_codes.txt", 'r') as f:
    codes = f.readlines()

# Clean and prepare data
codes = list(map(lambda x: x.strip(), codes))
datalist = []
for code in codes:
    datalist.append({
        "regionCategory": "State",  # Example: "County" or "Region"
        "region": "North Carolina Statewide",  # Example: "Wake" for Wake County
        "IndustryGroup2": code,  # NAICS 4-digit code, e.g., "1124" for Sheep and Goat Farming
        # "CompanyName": "",  # Optional: Add a company name or keyword
        # "stype": "c",  # Match type for CompanyName: 'c' (contains) or 's' (starts with)
        # "STFIPS": "37",  # State FIPS code for North Carolina
    })

# URL for form submission
url = "https://accessnc.nccommerce.com/BusinessSearch/Company/list"

# Process each data item with a progress bar
for data in tqdm(datalist, desc="Processing NAICS codes"):
    response = requests.post(url, data=data)
    html_parse = bs(response.text, 'html.parser')
    
    # Extract detail links
    for link in html_parse.find_all("a", string=lambda text: text == "Detail"):
        local_link = link.get("href")
        os.popen(f"echo {local_link} >> all_links.txt")
    
    # Respect rate limits
    time.sleep(0.05)
```

Store all full links

```python
all_links = []
base_url = "https://accessnc.nccommerce.com"
with open("all_links.txt","r") as f:
    all_links = f.readlines()
all_links_full = list(map(lambda x: base_url + x, all_links))
with open("all_links_full.txt","w") as f:
    f.writelines(all_links_full)
```

```python
with open("all_links_full.txt", 'r') as f:
    all_links_full = f.readlines()

# Clean and prepare data
all_links_full = list(map(lambda x: x.strip(), all_links_full))
link = all_links_full[0]
response = requests.get(link, headers=set_headers)
html_parse = bs(response.text, "html.parser")
for line in html_parse:
    print(line)
```

```python
# Locate the table
table = html_parse.find("table")

# Initialize an empty dictionary to store extracted data
data_dict = {}

# Extract table rows
for row in table.find_all("tr"):
    cells = row.find_all("td")
    if len(cells) == 2:  # Ensure there are two cells (label and value)
        label = cells[0].get_text(strip=True).replace(":", "")
        value = cells[1].get_text(strip=True).replace(" ","").replace("\n","").replace("\r","")
        data_dict[label] = value

# Display the extracted data
data_dict
```
